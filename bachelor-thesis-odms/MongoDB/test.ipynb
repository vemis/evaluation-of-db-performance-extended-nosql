```python
import pandas as pd
```

```python

customer_cols = [
    "c_custkey", "c_name", "c_address", "c_nationkey",
    "c_phone", "c_acctbal", "c_mktsegment", "c_comment"
]

supplier_cols = [
    "s_suppkey", "s_name", "s_address", "s_nationkey",
    "s_phone", "s_acctbal", "s_comment"
]

customer = pd.read_csv("..\\dataset\\TPC-H\\tpch-data\\customer.tbl", delimiter="|", header=None, names=customer_cols, usecols=range(len(customer_cols)))
supplier = pd.read_csv("..\\dataset\\TPC-H\\tpch-data\\supplier.tbl", delimiter="|", header=None, names=supplier_cols, usecols=range(len(supplier_cols)))

#customer = pd.read_csv("..\\dataset\\TPC-H\\tpch-data\\customer.tbl", delimiter="|")
#supplier = pd.read_csv("..\\dataset\\TPC-H\\tpch-data\\supplier.tbl", delimiter="|")
```

```python
# INTERSECT
# Get unique keys from each table
customer_keys = set(customer["c_custkey"])
supplier_keys = set(supplier["s_suppkey"])

# INTERSECT
common_keys = customer_keys.intersection(supplier_keys)

# Convert to DataFrame (optional)
result = pd.DataFrame(sorted(common_keys), columns=["key"])

print(result)
print("len: " + str(len(result)))
```

```python
# DIFFERENCE
customer_keys = set(customer["c_custkey"])
supplier_keys = set(supplier["s_suppkey"])
print(len(customer_keys))
print(len(supplier_keys))


difference_keys = customer_keys.difference(supplier_keys)

result = pd.DataFrame(sorted(difference_keys), columns=["key"])
print(result)
```
